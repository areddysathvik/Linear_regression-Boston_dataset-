```python
# applying scaling to the independent features so that the calculations happens faster

# diving into x and y
X = df.drop('target',axis=1)
y = df['target']
```

```python
# now apply sccaling to independent features

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X # scales data with normalizing them into mean = 0 and standard deviation of 1
```

```python
#creating the model

from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
# dividing into train and test sets to check accuracy

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape # we dont X as capital because it is a matrix and
# all matrixs should be represented as capitals
```

```python
#training the model

model.fit(X_train,y_train)
```

```python
# predicting the values using our model
y_pred = model.predict(X_test)
```

```python
y_test # true values
```

```python
y_pred # predicted values
```

```python
#checking accuracy

from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```
